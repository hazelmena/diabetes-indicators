```python
import pandas as pd
import hvplot.pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('modified.csv')
```

```python
gender_mapping = {0.0: 'Female', 1.0: 'Male'}
df['Sex'] = df['Sex'].map(gender_mapping)
```

```python
diabetes_indicator = {0.0: 'No Diabetes', 1.0: 'Has Diabetes'}
df['Diabetes_binary'] =df['Diabetes_binary'].map(diabetes_indicator)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Diabetes_binary', data=df)
plt.title('Distribution of Diabetes by Gender')
```

```python
plt.figure(figsize=(6, 6))
df['Diabetes_binary'].groupby(df['Sex']).value_counts().unstack().plot.pie(subplots=True, autopct='%1.1f%%', startangle=0, legend=False)
plt.title('Distribution of Diabetes by Gender')
```
